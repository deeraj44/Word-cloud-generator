```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
text = open('a80sup.txt', mode = 'r',encoding = 'utf-8').read()
stopwords = STOPWORDS
wc = WordCloud(
    background_color = 'white',
    stopwords = stopwords,
    height = 600,
    width = 400,
)
wc.generate(text)
wc.to_file('wordcloud1_output.png') #SAVES WORDCLOUD TO THE CURRENT DIRECTORY
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear') #DISPLAYS THE WORDCLOUD IN THE NOTEBOOK
plt.axis('off')
plt.show()
```
